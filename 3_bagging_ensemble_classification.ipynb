## Oversampling

```python
def oversampling(donnees,nom_var):
    data=donnees.copy()
    indexval1=data.index[data[nom_var]==1]
    indexval0=data.index[data[nom_var]==0]
    #tirage aléatoire avec remise des valeurs du groupe 1
    newliste=np.random.choice(indexval1,indexval0.shape[0],replace=True).astype('int32')
    newliste1=[]
    for val in newliste:
        newliste1.append(np.min(np.where(data.index==val)))
    data_over=data.take(newliste1)
    data_over=pd.concat([data_over,data.loc[indexval0]],axis=0)
    return(data_over)


train_set_over = oversampling(train_data[liste_des_variables],'assur_caravane')#initialisation et choix pour replacement qui permet de spécifier
test_set_over = oversampling(test_data[liste_des_variables],'assur_caravane')#initialisation et choix pour replacement qui permet de spécifier



X_train_over=train_set_over[variables_quanti] 
X_test_over=test_set_over[variables_quanti]

y_train_over=train_set_over['assur_caravane']
y_test_over=test_set_over['assur_caravane']
```

## **KNN** :

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train_set_over.drop(["assur_caravane"],axis=1))

X_train_cr = scaler.transform(X_train_over)
X_test_cr = scaler.transform(X_test_over)

df_train_cr=pd.concat([pd.DataFrame(X_train_cr,index=X_train_over.index),y_train_over],axis=1)
df_test_cr=pd.concat([pd.DataFrame(X_test_cr,X_test_over.index),y_test_over],axis=1)

X_train_over=df_train_cr.iloc[:,0:7]
X_test_over=df_test_cr.iloc[:,0:7]

y_train_over=df_train_cr['assur_caravane']
y_test_over=df_test_cr['assur_caravane']
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

y_train=y_train_over
y_test=y_test_over

k=20
knn=KNeighborsClassifier(n_neighbors=k).fit(X_train_over,y_train_over)
y_predict=knn.predict(X_train_over)
y_predict_test1=knn.predict(X_test_over)


print("KNN pour K=",k)
#calcule de précision-entrainnement
print("calcule de précision-entrainnement : ", knn.score(X_train_over, y_train_over))
#calcule de prediction-test
print("calcule de précision-test : ", knn.score(X_test_over, y_test))

print("\nMatrice de confusion avec LISTE train-validation :")
print(confusion_matrix(y_train,y_predict))

print("\nMatrice de confusion avec LISTE test-validation :")
print(confusion_matrix(y_test,y_predict_test1))

print(classification_report(y_test_over,y_predict_test1))
```

### Methode du vote (Pour la classification):

```python
def methode_vote(l):
    n=len(l)
    L=[]
    for i in range(len(l[0,:])):
        res=0
        for valeur in l[:,i]:
            res+=int(valeur)
        res=res/n
        if res>=0.5 :
            L.append(1)
        else :
            L.append(0)
    return L

l=np.array([y_predict_test1,y_predict_test2,y_predict_test3,y_predict_test4,y_predict_test5])
y_pred=methode_vote(l)
```
